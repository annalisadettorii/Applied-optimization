# Exercise 1, Analytic Optimization

Consider the minimization problem $$\min_{x\in\mathbb{R}}\left(1-\frac{1}{2}x^2\right)exp\left(-\frac{1}{2}x^2\right)$$

(a) Compute the first and second derivative of the objective function with respect to x.

The first derivative of the objective function is $$f'(x)=\frac{-4x+x^3}{2e^{\frac{x^2}{2}}}\text{,}$$

while its second derivative is $$f''(x)=\frac{7x^2-x^4-4}{2e^{\frac{x^2}{2}}}\text{.}$$

(b) Set the first derivative equal to zero and solve for x.

$f'(x)=0 $ if and only if $f'(x)=\frac{-4x+x^3}{2e^{\frac{x^2}{2}}}=0$ if and only if $-4x+x^3=0$ if and only if $x\in\{0,2,-2\}$ and we have that $f(0)=0,\; f(2)=-0.13,\; f(-2)=-0.13$.

(c) Inspect the second derivative of your solutions and identify local minima.

In this point we computed the values of the second derivative in the points where the gradient is null so that we can define if the found points are maxima or minima, because if $f''(x)<0$ we have a maximum and if $f''(x)>0$ we have a minimimum. This is what we found out:

- $f''(0)=-2<0$, so $f(0)=0$ is a maximum

- $f''(2)=0.54>0$, so $f(2)=-0.13$ is a minimum

- $f''(-2)=2.7>0$, so $f(-2)=-0.13$ is a minimum


(d) Check the limits of the domain against your local minima and thus verify that you found global minima.

$$\lim_{x\rightarrow +\infty}f(x)=\lim_{x\rightarrow +\infty}\left(1-\frac{1}{2}x^2\right)exp\left(-\frac{1}{2}x^2\right)=0$$

$$\lim_{x\rightarrow -\infty}f(x)=\lim_{x\rightarrow -\infty}\left(1-\frac{1}{2}x^2\right)exp\left(-\frac{1}{2}x^2\right)=0$$

So the minima are global, because the function is not decreasing towards infinity if we study the functions on the limits of the domain $[-\infty,+\infty]$.

The fact is that $f(\pm 2)<0$.

```python
# function f(x)
def f(x):
    return np.tanh((x + 1) * (x - 0.8) * (x + 0.1))

x = np.linspace(-10, 10, 1000)  
y = f(x)

# plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# derivative of f(x)
def f_derivative(x):
    sech_squared = 1 / np.cosh((x + 1) * (x - 0.8) * (x + 0.1))**2
    derivative = sech_squared * ((x - 0.8)*(x + 0.1) + (x + 1)*(2*x - 0.7))
    return derivative

# derivative in 0
f_derivative_at_0 = f_derivative(0)

t = np.linspace(-0.5, 0.5, 1000) 
d = (f(t) - f(0)) / t #d(t)

# plot
plt.figure(figsize=(8, 6))
plt.plot(t, d, label="d(t) = (f(t) - f(0)) / t", color='r')
plt.axhline(f_derivative_at_0, color='g', linestyle='--', label=f'$f\'(0)$ = {f_derivative_at_0}')
plt.xlabel('t')
plt.ylabel("d(t)")
plt.title("Plot of d(t) and f'(0)")
plt.grid(True)
plt.legend()
plt.show()
```

We can see that $d(t)$ for $t$ that goes to $\pm \infty$ never reaches the ax $d(t)=0$

For $t=0$ we have that the derivative is exactly equal to the limit of the different quotient in $t_0=0$ and this is what we expected, because the different quotient evalued in a point is the definition of derivative in that specific point.

```python
dimff = np.linspace(-10, 10, 1000)
X_f, Y_f = np.meshgrid(dimff, dimff)

Zf = f((X_f, Y_f)) 

def gradf(x0): #define gradient 
    dfdx=(1 / 50) * x0[0] + (3 / 2) * np.sin(3 * x0[0] -3 * x0[1])+ (3/2)* np.sin(3*x0[0]+ 3*x0[1])
    dfdy=(1 / 50) * x0[1] - (3 / 2) * np.sin(3 * x0[0] - 3 * x0[1])+ (3/2)* np.sin(3*x0[0]+ 3*x0[1])
    return np.array([dfdx,dfdy])

plt.pcolormesh(X_f, Y_f, Zf)
plt.colorbar()
plt.title("f(x,y) heatmap")
plt.scatter(x_0[0],x_0[1], color='r')
x_current=x_0
i=1 # number of iterations 
eps=0.1 # gradient norm tolerance
eta= 0.11 #stepsize 
max_iter=100 # upper bound on iterations
while np.linalg.norm(gradf(x_current)) > eps:  # while the the gradient norm > eps
    x_new= x_current - eta * gradf(x_current) # get new point
    plt.scatter(x_new[0],x_new[1], color='w') 
    x_current=x_new
    i+=1 
    if i > max_iter:
        break 
    
plt.xlim(-2,5)
plt.ylim(2.5,7)
```

Through this heatmap we can see graphically how the algorithm of the gradient descent converges to the closest minima starting from the point $x_0$, by following the direction for which the gradient points down. In such a way we found a minimum. We can not appreciate it graphically, but this is how it would look like in a 3D plot:

```python
#3d plot with the code of the last sheet

def f(xy): #defining the function f
    x, y = xy
    return 1/100 * (x**2 + y**2) - 1/2 * (np.cos(3*x - 3*y) +  np.cos(3*x + 3*y)) 

dimf = np.linspace(0, 5, 1000)
Xf, Yf = np.meshgrid(dimf, dimf)

Z_f = f((Xf, Yf)) 
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')


surf = ax.plot_surface(Xf, Yf, Z_f, cmap = 'plasma')

point_x = 0.1
point_y = 5
point_z = f((point_x, point_y))

ax.scatter(point_x, point_y, point_z, color='red', s=50, label='$x_0$')

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x_1,x_2)$')

plt.legend()
plt.show()
```

```python
#function of the gradient descent
def gradient_descent(x0,eta, eps=0.5, max_iter=1000000):
    x_current=x0
    i=1
    while  np.linalg.norm(gradf(x_current)) > eps: 
        x_new= x_current - eta * gradf(x_current)
        x_current=x_new
        i+=1 
        if i > max_iter:
            print (f'algorithm did not converge in {max_iter} iterations using {eta} as stepsize\n\
                   Coordinates are x={x_current[0]} and y={x_current[1]}\n\
                    ----------------------------------')
            return i
            
    
    print(f'algorithm converged in {i} iterations using {eta} as stepsize\n\
          Coordinates are x={x_current[0]} and y={x_current[1]}\n\
            ----------------------------------')
    return i
```

```python
#computing the iterations
etas = np.linspace(0.01,1,100)
etas_p=[]
iterations=[]
for eta in etas:
    iterations.append(gradient_descent(x_0, eta, eps=0.1,max_iter=100))
    etas_p.append(eta)
```

```python
#plotting the iterations for every step eta
plt.plot(etas_p, iterations)
plt.xlabel('$\eta$')
plt.ylabel('Number of iterations')
plt.axvline(x=0.22, color='red', linestyle='--', label='Local minima boundary')
plt.legend()

plt.show()
```

The gradient descent will always converge linearly. However, the choice of the amplitude of $\eta$ is critical, since taking a too big step may overshoot the closest local minima, while a too small one may need a too high number of steps to converge, even if it belongs to the attraction center of the closest local minima. This happens if the algorithm is implementing an upper bound on the maximum number of iterations, so it will stop if the convergencetakestoolong.

The approximation of $g(t)$ by a function of the second order can be done with the Taylor expansion and so we get: $$g(\eta)\approx 0.975858 + 1.73171 \eta - 2.4737 \eta^2$$

```python
# g(t)
def g(t):
    return 0.2501 + 0.00767 * t**2 + 0.08763 * t - 0.5 * (np.cos(-14.7 - 2.679 * t) + np.cos(15.3 + 2.577 * t))

# third-order expansion of g(t)
def g_third_order(t):
    return 0.975858 + 1.73171 * t - 2.4737 * t**2 

t_values = np.linspace(-1, 1, 400)

# values of g(t) and its third-order expansion
g_values = [g(t) for t in t_values]
g_third_order_values = [g_third_order(t) for t in t_values]


plt.figure(figsize=(8, 6))

# plot of g(t) and its expansion
plt.plot(t_values, g_values, label='g(t)')
plt.plot(t_values, g_third_order_values, label='g(t)\'s $3rd$ order expansion)', color='red')
plt.xlabel('t')
plt.ylabel('g(t)')
plt.title('g(t) and its third order expansion')
plt.legend()

plt.show()
```

By looking at the function and its approximation espansion in the interval $[-1,1]$ we can see that they are very close to each other. If we decide to approximate to the fourth order instead of the fifth it's even more clear that the function approximates better for every term we add to the Taylor expansion. You observe the population size $x(t)$ of an animal species on a group of islands over time $t$. The animals are not preyed upon by any predator and you regularly add a fixed number of individuals $c>0$ that is the same for each island.

for a growth rate $r\in\mathbb{R}$ and an addition amount $c$, we can approximate the population size for each island as a function of
time given by the following differential equation: $\frac{dx}{dt}=rx+c$ which has the solution $$x_{r,\;c,\;k}(t)=k\;exp(rt)-\frac{c}{r}\text{,}$$
where $k$ is an additional parameter.

Suppose you are given a data set that documents the time of observation $t$, the island ID $i$, and the number of individuals $x_i(t)$. Use this data to estimate the growth rate $r_i$ for each island, i.e. find $r_i,\;c,\;k_i$ such that $x_{r_i,\;c,\;k_i}(t)\approx x_i(t)$ for all $i$.

Keep in mind, that the model is only an approximation.

(a) Formalize the problem described above. Explain your choices! Note that the above description does not

To solve such a problem we decided to minimize the mean squared error between the observed population $x(t)$ and the estimated population $\tilde{x}(t)$ through a dataset provided. We are looking for the coefficients $r_i,\;k_i$ for $i\in\{1,.\;.\;, n\}$ where $n$ is the number of islands and $c$. As constraints we want that:
- the paramers $c_i$ to be positive, because they are the fixed number of regulary addedd animals for every island $i$;
- $\tilde{x}_j(t_j)>0$ to be positive, because we don't want a negative number of the population size estimated for $j$ considered sample;
- $\tilde{x}_j(t_j)=k_i\cdot exp(r_it_j)-\frac{c}{r_i}$, because we need that the population size to be such a function of these parameters

The choice of $MSE$ depends on the fact that it is differentiable, which is a fundamental property when using gradient-based optimization algorithms for parameter estimation. $MAE$ on the other hand, since makes use of absolute value, it's not differentiableeverywhere, it has problems with differentiability in the angolous points.

Our optimization problem in standard form is

$\min\limits_{r_i,\;k_i,\;c} \frac{1}{800}\sum\limits_{j=0}^{799}(x_j(t)-\tilde{x}_j(t))^2,\; i \in \{0,\;.\;.\;.\;, 7\}$

```python
import pandas as pd
#we gave a look to the dataset to understand how it's made
population_df = pd.read_csv('population.csv')
population_df.describe().T
```

```python
import warnings

warnings.filterwarnings("ignore", category=RuntimeWarning)

from scipy.optimize import minimize  
# population model function
def population_model(t, r, c, k):
    return k * np.exp(r * t) - c / r

# MSE
def mse_loss(params, t, observed_population):
    r, c, k = params
    predicted_population = population_model(t, r, c, k)
    mse = np.mean((observed_population - predicted_population) ** 2)
    return mse

estimated_parameters = []
num_islands = population_df['island_id'].unique()
island_ids = population_df['island_id']

for island_id in num_islands:  # for each id try 10 different starting points
    rs=[]
    cs=[]
    ks=[]
    mses=[]

    island_data_mask = (island_ids == island_id)  
    t_i = population_df['time'][island_data_mask] # get the ith time|island_id
    observed_population_i = population_df['population_size'][island_data_mask] # get the true population|island_id
    
    # initial parameter guess for the current island from a normal distribuition
    for i in range(10):
        init_r= np.random.normal(0,1)
        init_c=np.random.normal(0,1)
        init_k= np.random.normal(0,1)
        initial_guess=[init_r,init_c,init_k]
        
        # perform parameter estimation for the current island
        
        result = minimize(mse_loss, initial_guess, args=(t_i, observed_population_i), method='L-BFGS-B',
                          bounds=[(-1, 1e6), (1, 1e6), (-1e6, 1e6)])
        estimated_params_i = result.x
        
        rs.append(estimated_params_i[0])
        cs.append(estimated_params_i[1])
        ks.append(estimated_params_i[2])
        mses.append(result.fun)
        
        res=pd.DataFrame({'r':rs, 
                          'c':cs,
                          'k':ks, 
                          'mse':mses})
        
        best_mse= res.loc[res['mse'].idxmin()].values.tolist()

    estimated_parameters.append(best_mse[:-1])


# estimated_parameters for each island
    print(f'Island {island_id}: r = {best_mse[0]}, c= {best_mse[1]}, k={best_mse[2]}')
```

The code initializes an empty list <tt>estimated_parameters</tt> to store the estimated parameters for each island. It also identifies the unique island <tt>IDs</tt> in the dataset, <tt>num_islands</tt>, and extracts the island <tt>IDs</tt> and related data from the DataFrame, then it iterates through each island identified by its unique ID.

For each island, it estimates parameters by trying 10 different initial parameter guesses. For each initialization attempt, it randomly generates initial values for <tt>r</tt>, <tt>c</tt>, and <tt>k</tt> drawn from a standard normal. 

Using the initial parameter guesses, the code uses the minimize function from the SciPy library to optimize the parameters <tt>'(r, c, k)'</tt> for the population growth model. It employs the <tt>'-BFGS-B'</tt> method for bounded optimization and imposes bounds on the parameters.

For each initialization attempt, it collects the estimated values of <tt>r, c, k </tt>, and the resulting MSE in lists <tt>(rs, cs, ks, mses)</tt> specific to that island and it creates a DataFrame res to organize the results for each initialization attempt, with columns for <tt>r, c, k</tt>, and <tt>MSE</tt>.

After all initialization attempts, it identifies the initialization attempt with the lowest <tt>MSE</tt> for the current island.

Finally, the estimated parameters for the island with the lowest <tt>MSE</tt> are appended to the <tt>estimated_parameters</tt> list. The list contains the best estimates of <tt>r, c</tt>, and <tt>k</tt> for each island.

For each island, the code prints the best-fitting parameters <tt>(r, c, k)</tt> and the correspondingisland<tt>ID</tt>.
