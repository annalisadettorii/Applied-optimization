# 1. Minimum Enclosing Gift Box

Assume you wish to put a christmas gift into a round box. The corners of your gift item are given by a set of two-dimensional points $\mathbf{x}_1,\;.\;.\;. \;,\mathbf{x}_m\in\mathbb{R}^2$. Now, you wish to find the smallest round box such that your gift still fits inside.

We can formalize this problem as follows. Let $c$ be the location of the center of our box and let $r^2$ be the squared radius. Then we wish to solve the problem:
$$\min_{\mathbf{c}\in\mathbb{R}^2,\;r^2\in\mathbb{R}}r^2$$
$$s.t.\;(\mathbf{c}-\mathbf{x}_i)^T\cdot(\mathbf{c}-\mathbf{x}_i)\leq r^2\;\;\;\;\;\;\;\;\;\;\;\forall i \in\{1,\;.\;.\;.\;,m\}$$

(a) Write down the Lagrange dual of this problem.

**Definition**: For a convex primal problem

$$\min_{\mathbf{x}\in\mathbb{R}^2}f(\mathbf{x})$$

$$s.t.\;g_i(\mathbf{x})\geq 0$$

the Wolfe dual is defined as

$$\sup_{\mathbf{x}\in\mathbb{R}^2,\;\lambda\in\mathbf{R}^m}f(\mathbf{x})-\sum_{i}(\lambda)_ig_i(\mathbf{x})$$
$$s.t.\;(\lambda)_i\geq 0$$
$$\nabla_{\mathbf{x}}f(\mathbf{x})=\sum_{i=1}^{m}(\lambda)_i\nabla_{\mathbf{x}}g_i(\mathbf{x})$$


Let's first compute the Lagrangian form as follows:
$$ \mathcal{L}(r^{2},\mathbf{c}, \mathbf{\lambda}) = r^{2} + \sum_{i=1}^{m} \lambda_i \left((\mathbf{c}-\mathbf{x}_i)^{T}(\mathbf{c}-\mathbf{x}_i)-r^{2}\right)$$

from which the Lagrangian dual is defined as follows:


$$ \sup_{\lambda\in\mathbb{R}^{m}} \min_{r^{2},\mathbf{c}}\mathcal{L}(r^{2},\mathbf{c}, \mathbf{\lambda})$$
$$\;\;\;\;\;\;\;s.t. \;\lambda_i \geq 0 \;\;\;\;\;\;\;\; \forall i \in\{1,\;.\;.\;.\;,m\}$$

(b) Write down the Wolfe dual of this problem.

To compute the Wolfe dual we need to start again from the Lagrangain dual, obtaining the following setting:
$$ \sup_{r^{2}, \mathbf{c},\mathbf{\lambda} } \; r^{2} + \sum_{i=1}^{m} \lambda_i \left((\mathbf{c}-\mathbf{x}_i)^{T}(\mathbf{c} -\mathbf{x}_i)-r^{2}\right)$$
$$\;\;\;\;\;\;\;s.t. \;\lambda_i \geq 0 \;\;\;\;\;\;\;\; \forall i \in\{1,\;.\;.\;.\;,m\}$$ 
$$\;\;\;\;\;\;\;\;\;\nabla_{r^{2},\mathbf{c}}\;r^{2} + \sum_{i=1}^{m} \lambda_i \left((\mathbf{c}-\mathbf{x}_i)^{T}(\mathbf{c}-\mathbf{x}_i)-r^{2}\right)=0$$

From the constraints we obtain the following:

$$\nabla_{r^{2}}\mathcal{L}(r^{2},\mathbf{c}, \mathbf{\lambda})= 1-\sum_{i=1}^{m}(\lambda)_i$$
$$\nabla_{\mathbf{c}}\;\mathcal{L}(r^{2},\mathbf{c}, \mathbf{\lambda})=2 \sum_{i=1}^{m} \lambda_i (\mathbf{c}-\mathbf{x}_i)$$

(c) Identify a feasible solution for $c$, depending on $\lambda$, using the Wolfe dual.

To find a feasible solution for $\mathbf{c}$ we can set $\nabla_{\mathbf{c}} \mathcal{L} = 0$, so we can basically re-use te constraint already treated in the previous point.  

$$\nabla_{\mathbf{c}} \mathcal{L} = 2 \sum_{i=1}^{m} \lambda_i (\mathbf{c}-\mathbf{x}_i)$$
By solving for $c$ we obtain:
$$c = \frac{\sum_{i=1}^{m}\lambda_i \mathbf{x}i}{\sum{i=1}^{m}\lambda_i } $$

(d) Re-write the Wolfe dual, plugging your results for the side constraints into the objective function. Simplify as far as possible.

$$ \sup_{r^{2},\mathbf{\lambda} } \; r^{2} + \sum_{i=1}^{m} \lambda_i \left((\frac{\sum_{j=1}^{m}\lambda_j\mathbf{x}i}{\sum{j=1}^{m}\lambda_j }-\mathbf{x}i)^{T}(\frac{\sum{j=1}^{m}\lambda_j \mathbf{x}i}{\sum{j=1}^{m}\lambda_j } -\mathbf{x}_i)-r^{2}\right)$$
$$s.t. \;\lambda_i \geq 0 \;\;\;\;\;\;\;\;  $$
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; \sum_{i=1}^{m}(\lambda)_i=1 \;\;\;\;\;\; \forall i \in\{1,\;.\;.\;.\;,m\}$

By plugging in the constraint we obtain the following:
$$\sup_{r^{2},\mathbf{\lambda} } \; r^{2}-\sum_{i=1}^{m} \lambda_i r^{2}$$
$$ s.t. \;\lambda_i \geq 0 \;\;\;\;\;\;\;\;\forall i \in\{1,\;.\;.\;.\;,m\} $$
leading to :
$$\sup_{\mathbf{\lambda} } \;0 $$
$$ s.t. \;\lambda_i \geq 0 \;\;\;\;\;\;\;\;  \forall i \in\{1,\;.\;.\;.\;,m\} $$

# 2 Reverse Modeling

Let $\mathbf{a},\;\mathbf{b},\;\mathbf{c},\;\mathbf{d},\;\mathbf{x}_1,\;.\;.\;.\;,\mathbf{x}_n\in\mathbb{R}^N,\;N\geq 2,\; \geq 1$ be real vectors. Consider the following optimization problems:

$$\min_{\alpha,\beta\in\mathbb{R}}||\alpha\mathbf{a}-\beta\mathbf{c}+(1-\alpha)\mathbf{b}-(1-\beta)\mathbf{d}|| \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{(a)} $$
$$s.t.\;0\leq\alpha,\beta\leq1$$

$$\min_{\xi_{1},\;.\;.\;.\;,\xi_{n}\in\mathbb{R}}\sum_{i=1}^{n}\left|\left|\mathbf{x}_i-\sum_{j=1}^{n}\xi_j\mathbf{x}_j\right|\right|^2  \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{(b)} $$
$$s.t.\;\sum_{i=1}^{n}\xi_i=1$$
$$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\xi_i\geq 0 \;\;\;\;\;\;\;\;\;\;\;\;\;\;\forall i \in\{1,\;.\;.\;.\;,n\}$$

$$\max_{\xi_{1},\;.\;.\;.\;,\xi_{n}\in\mathbb{R}}\sum_{i=1}^{n}\xi_i \left|\left|\mathbf{x}_i-\frac{1}{n}\sum_{j=1}^{n}\mathbf{x}_j\right|\right| \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{(c)} $$
$$s.t.\;\sum_{i=1}^{n}\xi_i=1$$
$$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\xi_i\geq 0 \;\;\;\;\;\;\;\;\;\;\;\;\;\;\forall i \in\{1,\;.\;.\;.\;,n\}$$

$$\min_{\mathbf{n}\in\mathbb{R}^N,\;d,\;\xi_{1},\;.\;.\;.\;,\xi_{n}\in\mathbb{R}} \sum_{i=1}^{n}\xi_i\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{(d)} $$
$$s.t.\; ||\mathbf{n}||=1$$
$$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\mathbf{n}^T\mathbf{x}_i-\xi_i=d \;\;\;\;\;\;\;\;\;\;\;\forall i \in\{1,\;.\;.\;.\;,n\}$$
$$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\xi_i\geq 0 \;\;\;\;\;\;\;\;\;\;\;\;\;\;\forall i \in\{1,\;.\;.\;.\;,n\}$$

$$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;s.t.\; \mathbf{w}^T\mathbf{x}_i\geq 1 \;\; \forall i \in\{1,\;.\;.\;.\;,n\}$$
$$\mathbf{w}^T\mathbf{p}=1$$

# 3 Hard to Optimize

Consider the following optimization problems:

$$\min_{(x,y)\in\mathbb{R}^2}\frac{1}{100}(x^2+y^2)-\frac{1}{2}(cos(3x-3y)+cos(3x+3y)) \;\;\;\;\;\;\;\text{(a)}$$

$$\min_{(x,y)\in\mathbb{R}^2} (x-1)^2+100(x^2-y)^2 \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{(b)}$$

$$\min_{(x,y)\in\mathbb{R}^2} xy \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\text{(c)}$$
$$s.t.\; x^2+y^2\leq 1+\frac{1}{5}cos\left(8arctan\left(\frac{x}{y}\right)\right)$$

For each of the problems _(a)_,_(b)_, and _(c)_...

(a)  ... name at least two algorithms from the lecture that can be applied to the problem.

_Problem (a)_: This is an unconstraint problem, so we can use without any problems algorithms such as the downhill simplex or the gradient descent method.

_Problem (b)_: This problem too is unconstrained and it's important to focus on the fact that the objective function is convex, so we can use the Newton method wich converges very quickly to convex problems. We can also use quasi-Newton methods to solve this problems in order not to compute the Hessian and the derivate of the function $f$.

_Problem (c)_: This problem has a constraint, so we must use methods for constrainted problems such as the Lagrange dual and the Wolfe dual.

(b) ... empirically evaluate the method. Answer the following questions (for different starting points):

- How well does the method identify the global optimum. (e.g. difference between found and global optimum
in terms of function value or position in space; to do so first find the analytical solution)
- How long does it take to do so. (e.g. runtime, number of iterations, or number of function/gradient evaluations)

```python
import pandas as pd
import numpy as np
import random
import time
from scipy.optimize import minimize

def objective_function_1(params): # first function
    x, y = params
    return 1/100 * (x**2 + y**2) - np.cos(3*x) * np.cos(3*y)

methods1=['CG', 'BFGS', 'Nelder-Mead'] # methods

df1=pd.DataFrame(columns=['method','avg_time', 'avg_n of iterations','convergence_prop', 'best_function_value']) #info about the methods

n=1000

for method in methods1:
    d={'method':method,'avg_time':0, 'avg_n of iterations':0,'convergence_prop':0 , 'best_function_value':[]} 
    for i in range(n):
        
        start=[random.random(),random.random()] # random point as starting point 
        time_start= time.time()
        result = minimize(objective_function_1, start, method=method)
        time_end = time.time()
        runtime = time_end-time_start
        d['avg_time']+=(runtime/n) 
        d['avg_n of iterations']+=(result.nit/n)
        converged = 1 if result.success else 0
        d['convergence_prop']+=(converged/n)
        d['best_function_value'].append(result.fun)

    d['best_function_value']=min(d['best_function_value'])
    
    df1.loc[len(df1)] = d # updated new line

    

df1
```

```python
def objective_function_2(params):
    x, y = params
    return (x - 1)**2 + 100 * (x**2 - y)**2

def jacobian(params):
    x, y = params
    df_dx = 2 * (-1 + x * (1 + 200 * (x**2 - y)))
    df_dy = -200 * (x**2 - y)
    return np.array([df_dx, df_dy])

methods2 = ['CG', 'BFGS', 'Newton-CG']
df2 = pd.DataFrame(columns=['method', 'avg_time', 'avg_n of iterations', 'convergence_prop', 'best_function_value'])

n = 1000

for method in methods2:
    d = {'method': method, 'avg_time': 0, 'avg_n of iterations': 0, 'convergence_prop': 0, 'best_function_value': []}
    for i in range(n):
        start = [random.random(), random.random()]  # random point as starting point
        time_start = time.time()

        if method == 'Newton-CG':
            result = minimize(objective_function_2, start, method=method, jac=jacobian)
        else:
            result = minimize(objective_function_2, start, method=method)

        time_end = time.time()
        runtime = time_end - time_start
        d['avg_time'] += (runtime / n)
        d['avg_n of iterations'] += (result.nit / n)
        converged = 1 if result.success else 0
        d['convergence_prop'] += (converged / n)
        d['best_function_value'].append(result.fun)

    d['best_function_value'] = min(d['best_function_value'])
    df2.loc[len(df2)] = d

df2
```

```python
def objective_function_3(params):
    x, y = params
    return x * y

def constraint(params):
    x, y = params
    return 1 + 1/5 * np.cos(8 * np.arctan(x/y)) - x**2 - y**2

methods3 = ['COBYLA', 'SLSQP']
df3 = pd.DataFrame(columns=['method', 'avg_time', 'avg_n of iterations', 'convergence_prop', 'best_function_value'])

n = 1000

for method in methods3:
    d = {'method': method, 'avg_time': 0, 'avg_n of iterations': 0, 'convergence_prop': 0, 'best_function_value': []}
    for i in range(n):
        start = [random.random(), random.random()]
        time_start = time.time()
        
        result = minimize(objective_function_3, start, method=method, constraints={'type': 'ineq', 'fun': constraint})
        
        time_end = time.time()
        runtime = time_end - time_start
        
        d['avg_time'] += (runtime / n)
        d['avg_n of iterations'] += (result.nfev / n)  # Use 'nfev' for COBYLA
        converged = 1 if result.success else 0
        d['convergence_prop'] += (converged / n)
        d['best_function_value'].append(result.fun)

    d['best_function_value'] = min(d['best_function_value'])
    df3 = df3.append(d, ignore_index=True)

df3
```
